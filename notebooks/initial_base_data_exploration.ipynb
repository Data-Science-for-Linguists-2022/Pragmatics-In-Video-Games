## The Elder Scrolls Data:

```python
# It is stored as a JSON file
elder_scrolls_df = pd.read_json(filepath + "/The Elder Scrolls/data/imperial_library_20200626.json")
display(elder_scrolls_df.head())

# To quote Ross Geller: "PIVOT, PIVOT, PIVOT"
elder_scrolls_df = elder_scrolls_df.transpose()
display(elder_scrolls_df.head())
```

```python
# Now let's get the names into their own column
elder_scrolls_df['title'] = [re.sub(r'https://www.imperial-library.info/content/', '', title) for title in elder_scrolls_df.index]

# Create a new DataFrame to get rid of the weirdness
elder_scrolls_df = elder_scrolls_df[['title', 'author', 'description', 'game', 'text', 'url']]
elder_scrolls_df = elder_scrolls_df.reset_index(drop=True)
display(elder_scrolls_df.head(20))
```

```python
# Let's get some info about our cleaned data
print('===================== ELDER SCROLLS DATAFRAME =====================')
print(elder_scrolls_df.info())
print(elder_scrolls_df.shape)
# I COUNT ANONYMOUS ENTRIES AS 1
print("UNIQUE AUTHORS:", len(set(elder_scrolls_df['author'])))
print("UNIQUE ENTRIES", len(set(elder_scrolls_df['title']))) # Good, no dupes
```

```python
# Let's make sure the list information was stored correctly
print(type(elder_scrolls_df.loc[0, 'game'])) # WE'RE GOOD!!!

# Let's get a master list of all the games featured
master_list = list()

for glist in elder_scrolls_df.loc[:, 'game']:
    for game in glist:
        if game not in master_list:
            master_list.append(game)
print(master_list)
print(len(master_list))
```

```python
# Now let's try to clean up the text data
print(elder_scrolls_df.loc[19, 'text'][:300]) # Looks like that's not an issue, I replaced this line with various rows
```

```python
# Let's make a word count section to our dataframe
# I decided to play it safe and use nltk.word_tokenize since, looking at the JSON, the data's been thoroughly kept already. 
# It's also nice to use an already established parsing-standard.
elder_scrolls_df['word_count'] = elder_scrolls_df.text.map(lambda x : len(nltk.word_tokenize(x)))

# Reorganize... again
elder_scrolls_df = elder_scrolls_df[['title', 'author', 'description', 'game', 'text', 'word_count', 'url']]
display(elder_scrolls_df.head(20))
```

```python
# As a data sample, let's give the person any text entries where the text is <= 50 words long. I want the viewer to be more inclined to go to the original dataset
# Creating the sample DataFrame
elder_sample_df = elder_scrolls_df[elder_scrolls_df['word_count'] <= 50]
# Let's get some info about our sampled data
print('===================== ELDER SAMPLE DATAFRAME =====================')
print(elder_sample_df.info())
print(elder_sample_df.shape)
# I COUNT ANONYMOUS ENTRIES AS 1, NOTE: NOT EVERYTHING HAS A DESCRIPTION!!!
print("UNIQUE AUTHORS:", len(set(elder_sample_df['author'])))
print("UNIQUE ENTRIES", len(set(elder_sample_df['title']))) # Good, no dupes

display(elder_sample_df.tail(20))
```

```python
# Finally, let's pickle both the main elder_scrolls_df DataFrame and elder_sample_df DataFrame and store them in the appropriate locations
# First the main one
elder_scrolls_df.to_pickle('../private/pickled_dfs/elder_scrolls.pkl')
elder_sample_df.to_pickle('../sample_data/elder_sample.pkl')
```

# Torchlight Data:

**PLEASE NOTE** This data is stored in both a `.csv` and a `.json`. Please see [orignal documentation](https://github.com/hmi-utwente/video-game-text-corpora/tree/master/Torchlight%20II) for more information (no, the script doesn't work, I tried).
```python
torchlight_df = pd.read_csv(filepath + '/Torchlight II/data/dataset_200630.csv')
display(torchlight_df.head())
```

```python
# Get rid of some files which relate
# to the game engine. Also remove raw text, add word count, and reorganize columns.
torchlight_df['word_count'] = torchlight_df['text'].map(lambda x: len(nltk.word_tokenize(x)))
torchlight_df = torchlight_df[['speaker', 'text', 'word_count', 'dialogtype', 'quest_displayname', 'quest_name']]

display(torchlight_df.head())
print(torchlight_df.info())
```

```python
# Now, let's replace the NaN's in speaker
torchlight_df['speaker'].replace(to_replace=np.nan, value='NO SPEAKER', inplace=True)
display(torchlight_df.tail())
print(torchlight_df.info())
```

```python
# Let's see the NaNs for quest_displayname
# torchlight_df[torchlight_df.quest_displayname == np.nan] FOR SOME REASON, THIS DID NOT WORK
display(torchlight_df[torchlight_df.quest_displayname.isnull()])
# This is regular dialogue or background dialogue, just put 'NON-QUEST' here. They still have
# a quest name because they are still triggered by or associated with quests (e.g. character
# won't say something until after a quest is complete or event happened)

torchlight_df['quest_displayname'].replace(to_replace=np.nan, value='NON-QUEST', inplace=True)
display(torchlight_df.loc[901:905])
```

```python
# Make code sample the first... let's say 100 (it's a small dataset to begin with and I want to encourage people to go to the og repo)
torchlight_sample_df = torchlight_df.loc[:100]
torchlight_sample_df.to_pickle('../sample_data/torchlight_sample.pkl')
torchlight_df.to_pickle('../private/pickled_dfs/torchlight.pkl')
```

## Star Wars: Knights of The Old Republic

```python
# There's two csv's: one for the actual dialogue and other data, the other which contains additional metadata about the animations used with the dialogue
kotor_df = pd.read_csv(filepath + '/Star Wars: Knights of the Old Republic/data/dataset_20200716.csv')
meta_df = pd.read_csv(filepath + '/Star Wars: Knights of the Old Republic/data/dialoganimations.2da.csv')

display(kotor_df.head(8))
display(meta_df.head(8)) # ???????????????????? WHY ????????????????????????
```

```python
# Let's focus on the metadata DataFrame, let's drop it like it's hot
meta_df.dropna(axis='rows', inplace=True)
meta_df = meta_df.reset_index(drop=True)
display(meta_df.head(8))
print(meta_df.shape)
```

```python
# print(kotor_df.info())
# Let's drop audiofile and source_dlg since I don't have the original copy
# I am also going to replace NaNs in 'listener' and 'comment' with 'NO LISTENER' and 'NO COMMENT' respectively
kotor_df['listener'].replace(to_replace=np.nan, value='NO LISTENER', inplace=True)
kotor_df['comment'].replace(to_replace=np.nan, value='NO COMMENT', inplace=True)

# Finally, word count
kotor_df['word_count'] = kotor_df['text'].map(lambda x: len(nltk.word_tokenize(x)))
kotor_df = kotor_df[['speaker', 'listener', 'text', 'word_count', 'animation', 'next', 'previous', 'comment']]
display(kotor_df.head())
print(kotor_df.info())
```

```python
# Save a sample pkl and the main pkl, let the sample pkl be just the first 5000 data entries
kotor_sample_df = kotor_df[:5000]
kotor_sample_df.to_pickle('../sample_data/kotor_sample.pkl')
kotor_df.to_pickle('../private/pickled_dfs/kotor.pkl')
meta_df.to_pickle('../private/pickled_dfs/meta_kotor.pkl')
```

## Hollow Knight
This is the `.csv` I made from my HDialogueParser.py using techniques gathered from [*Fantastic Strings and Where to Find Them: The Quest for High-Quality Video Game Text Corpora*](https://judithvanstegeren.com/assets/2008-vanstegeren2020fantastic-preprint.pdf). I couldn't save it as a `.pkl` there. I added more manual tagging using regex, details in the [overview notebook](https://github.com/Data-Science-for-Linguists-2022/Sociolinguistics-In-Video-Games/blob/main/notebooks/overview.ipynb).

```python
# Get the csv
hollow_df = pd.read_csv('../private/self-made/hollow_knight.csv', index_col='Unnamed: 0')
# Let's try to remove what's in the tags to get more accurate data. Also rename dialogue to text
hollow_df['text'] = hollow_df['dialogue'].map(lambda x: re.sub(r'<description>.*</description>', '', x))
# Now word count
hollow_df['word_count'] = hollow_df['text'].map(lambda x: len(nltk.word_tokenize(x)))

# Organize
hollow_df = hollow_df[['characters', 'text', 'word_count']]
hollow_df.rename(columns={'characters': 'character'}, inplace=True)

display(hollow_df.head(3))
print(hollow_df.info())
```

```python
# Let's try to save it as a pkl, and also get a sample for the sample_data folder
hollow_sample = hollow_df # Since I made this dataset myself, I'll give it all!
hollow_sample.to_pickle('../sample_data/hollow_sample.pkl')
hollow_df.to_pickle('../private/pickled_dfs/hollow_knight.pkl') # ... Don't know WHAT changed, but okay...
```
